# 4- Execute Python SQL Script from input data and return output

```python
import pandas as pd

data = pd.read_csv("C:\Scripts\Ch 4 - Machine Learning Services\Sales.csv")
data.head()
```

```python
table = pd.crosstab(
    [data.Year, data.Client], # group by in rows
    data.Quarter, # group by in columns
    values = data.Revenue, # values to aggregate
    aggfunc= sum,
    margins= True
)

table.reset_index(inplace=True)

print(table)
```

# Detect Faces

```python
import cv2
```

```python
image_folder = r"C:\Scripts\Ch 4 - Machine Learning Services\images"
image_folder_detect = r"C:\Scripts\Ch 4 - Machine Learning Services\images\Detected"
img = r"C:\Scripts\Ch 4 - Machine Learning Services\images\\people.jpg"
```

```python
def detect(imagePath):
    image = cv2.imread(imagePath)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
#     cascPath = "haarcascadexml\\haarcascade_frontalface_default.xml"
    cascPath = r"C:\Scripts\Ch 4 - Machine Learning Services\\haarcascade_frontalface_default.xml"
    
    cascade = cv2.CascadeClassifier(cascPath)
    
    faces = cascade.detectMultiScale(
        gray,
        scaleFactor=1.1,
        minNeighbors=10,
        minSize=(30, 30),
        flags = cv2.CASCADE_SCALE_IMAGE
    )
    for (x, y, w, h) in faces:
        cv2.rectangle(image, (x, y), (x+w, y+h), (127, 255, 0), 8)
        cv2.imwrite(imagePath.replace("images\\","images\\Detected\\"), image);
```

```python
detect(img)
```
